```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
files = [
    "Streaming_History_Audio_2016-2020_0.json",
    "Streaming_History_Audio_2020-2021_1.json",
    "Streaming_History_Audio_2021-2023_2.json",
    "Streaming_History_Audio_2023-2024_3.json",
    "Streaming_History_Audio_2024-2025_4.json",
    "Streaming_History_Audio_2025_5.json",
]

dfs = [pd.read_json(f) for f in files]

data = pd.concat(dfs, ignore_index=True)

print("Total rows: ", len(data))
```

```python
data.head()
```

```python
data = data.rename(columns={
    "ts": "timestamp",
    "master_metadata_track_name": "track",
    "master_metadata_album_artist_name": "artist",
    "master_metadata_album_album_name": "album",
})

data.head()
```

```python
data["timestamp"] = pd.to_datetime(data["timestamp"], errors="coerce")

data["year"] = data["timestamp"].dt.year
data["month"] = data["timestamp"].dt.month
data["day"] = data["timestamp"].dt.day
data["hour"] = data["timestamp"].dt.hour
data["weekday"] = data["timestamp"].dt.day_name()

data.head()
```

```python
music_data = data.dropna(subset=["artist", "track"])

if "episode_name" in data.columns:
    podcast_data = data[data["episode_name"].notna()]
else:
    podcast_data = data[data["artist"].isna()]

podcast_data.head()
```

```python
sensCols = ["ip_addr", "conn_country"]
for col in sensCols:
    if col in music_data.columns:
        music_data = music_data.drop(columns=col)
    if col in podcast_data.columns:
        podcast_data = podcast_data.drop(columns=col)

music_data = music_data[music_data["ms_played"] > 0].reset_index(drop=True)

music_data.to_csv("spotify_music_history.csv", index=False)
podcast_data.to_csv("spotify_podcast_history.csv", index=False)
```

```python
music = pd.read_csv("spotify_music_history.csv")
podcasts = pd.read_csv("spotify_podcast_history.csv")

print ("Music rows:", len(music))
print ("Unique artists:", music["artist"].nunique())
print ("Unique tracks:", music["track"].nunique())

print ("\nPodcast rows:", len(podcasts))
if "episode_name" in podcasts.columns:
    print("Unique shows/episodes:", podcasts["episode_name"].nunique())
```
